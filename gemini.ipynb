```python
def run_vending_machine():
    products = {
        "1": {"name": "콜라", "price": 1000, "stock": 5},
        "2": {"name": "사이다", "price": 1000, "stock": 5},
        "3": {"name": "환타", "price": 900, "stock": 3},
        "4": {"name": "웰치스", "price": 1200, "stock": 4},
        "5": {"name": "옥수수수염차", "price": 1500, "stock": 2},
    }
    
    balance = 0

    print("--- 파이썬 자판기 ---")

    while True:
        print("\n[메뉴판]")
        for code, item in products.items():
            print(f"{code}. {item['name']} - {item['price']}원 (남은 수량: {item['stock']}개)")
        print(f"현재 투입 금액: {balance}원")
        print("--------------------")

        choice = input("돈을 넣어주세요 (100, 500, 1000원 단위) 또는 상품 번호를 입력하세요 (q: 종료, r: 잔돈 반환): ").lower()

        if choice == 'q':
            print("자판기를 종료합니다.")
            if balance > 0:
                print(f"남아있는 잔돈 {balance}원을 반환합니다.")
            break
        elif choice == 'r':
            if balance > 0:
                print(f"잔돈 {balance}원을 반환합니다.")
                balance = 0
            else:
                print("반환할 잔돈이 없습니다.")
            continue

        try:
            # 돈 투입
            if choice.isdigit():
                amount = int(choice)
                if amount in [100, 500, 1000]:
                    balance += amount
                    print(f"{amount}원 투입. 현재 잔액: {balance}원")
                elif choice in products:
                    # 상품 선택
                    selected_product = products[choice]
                    if selected_product["stock"] > 0:
                        if balance >= selected_product["price"]:
                            balance -= selected_product["price"]
                            selected_product["stock"] -= 1
                            print(f"{selected_product['name']}이(가) 나왔습니다. 남은 잔액: {balance}원")
                        else:
                            print(f"잔액이 부족합니다. {selected_product['price'] - balance}원이 더 필요합니다.")
                    else:
                        print(f"{selected_product['name']}은(는 품절입니다. 다른 상품을 선택해주세요.")
                else:
                    print("유효하지 않은 입력입니다. 100, 500, 1000원 또는 상품 번호를 입력해주세요.")
            else:
                print("유효하지 않은 입력입니다. 숫자를 입력하거나 'q', 'r'을 입력해주세요.")

        except ValueError:
            print("잘못된 입력입니다. 숫자를 입력해주세요.")

# 자판기 실행
if __name__ == "__main__":
    run_vending_machine()
```
